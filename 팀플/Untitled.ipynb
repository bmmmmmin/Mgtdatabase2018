```python
from pylab import plot, show
```

```python
# y축에 손님 질문지 각 항목의 평균값을 넣는다.
y = [3.64,3.88, 4.28, 4.12, 4.04, 3.76, 4, 3.8]
plot(y)
show
```

```python
# x축은 항목 이름
x= ["첫인사", "끝인사", "말투", "성실성", "설명", "맛 일정", "비품", "토핑바 청결"]
plot(x,y)
show
```

```python
# 그래프의 x축, y축, 제목의 이름을 붙임
from pylab import title, xlabel, ylabel
title('Twosome')
xlabel('Question')
ylabel('Degree')
plot(x,y)
show
```

```python
# 손님과 직원의 만족도 차이를 그래프로 알아보기 위함
from pylab import legend
고객 = [3.64,3.88, 4.28, 4.12, 4.04, 3.76, 4.00, 3.8]
점원= [3.56,3.56,3.56,3.78,3.22,3.78,3.44,3.33]
plot(x, 고객, x, 점원, marker='o')
title('Twosome')
xlabel('Question')
ylabel('Degree')
legend(['고객', '점원',], loc='best')
show()
```
